```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=2500,
    chunk_overlap=200,
    length_function=len,
    separators=["\n\n", "\n", " ", ""]
)
chunks = text_splitter.split_documents(pages)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
def get_embedding():
    embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
    return embeddings

embedding_function = get_embedding()
```

```python
test_vector = embedding_function.embed_query("Man")
print(test_vector)
print(len(test_vector))
```

```python
def similarity(a, b):
    a_vec = embedding_function.embed_query(a)
    b_vec = embedding_function.embed_query(b)
    return np.dot(a_vec, b_vec) / (np.linalg.norm(a_vec) * np.linalg.norm(b_vec))

print(similarity("Paris is the capital of France", "France's capital is Paris"))
```

```python
import chromadb
chromadb.__version__
```

```python
import uuid
def create_vector_DB(chunks, embedding_function , path):
  ids = [str(uuid.uuid5(uuid.NAMESPACE_DNS, doc.page_content)) for doc in chunks]

  unique_ids = []
  unique_chunks = []

  for doc, id in zip(chunks,ids):
    if id not in unique_ids:
      unique_ids.append(id)
      unique_chunks.append(doc)
  chunks = unique_chunks
  vectorstore = Chroma.from_documents(documents=chunks,
                                      ids=unique_ids,
                                      embedding = embedding_function,
                                      persist_directory=path)
  vectorstore.persist()
  return vectorstore
```

```python
vector_database = create_vector_DB(chunks, embedding_function , "chroma_vector_database")
```

```python
retriever = vector_database.as_retriever(search_type="similarity")
relevant_chunks = retriever.invoke("What does GPT-4 say about difference in pain threshold?")
relevant_chunks
```

```python
PROMPT_TEMPLATE = """
You are an assistant for question-answering tasks. Use the following pieces of retrieved context to answer
the question. If you don't know the answer, say that you don't know. DON'T MAKE UP ANYTHING.

{context}

---

Answer the question based on the above context: {question}
"""
```

```python
context_text =  "\n\n---\n\n".join([doc.page_content for doc in relevant_chunks])
prompt_template = ChatPromptTemplate.from_template(PROMPT_TEMPLATE)
prompt = prompt_template.format(context = context_text, question = "What does GPT-4 say about difference in pain threshold?")
print(prompt)
```

```python
llm.invoke(prompt)
```

```python
def format_docs(docs):
  return "\n\n---\n\n".join([doc.page_content for doc in docs])

class ExtractedInfo(BaseModel):
  paper_title: str = Field(description="The title of the paper")
  paper_summary: str = Field(description="A summary of the paper")
  publication_date: str = Field(description="The date the paper was published")
  authors: str = Field(description="The authors of the paper")
```

```python
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt_template
    | llm.with_structured_output(ExtractedInfo, strict=True)

)
```

```python
structured_response = rag_chain.invoke("Give me the paper title, paper summary, publication date and authors of the research paper")
```
